```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('netflix_titles.csv')
df.head()
```

```
df.info()
df.isnull().sum()
```

```
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna(df['rating'].mode()[0], inplace=True)
df.drop_duplicates(inplace=True)

df.isnull().sum()
```

### Data Cleaning Explanation
Missing values in the country column were filled with 'Unknown'.
Missing values in rating were filled using the mode.
Duplicate rows were removed to improve data quality.

```

sns.countplot(x='type', data=df)
plt.title("Movies vs TV Shows")
plt.show()

df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Countries")
plt.show()

sns.histplot(df['release_year'], bins=30)
plt.title("Release Year Distribution")
plt.show()

df['rating'].value_counts().plot(kind='bar')
plt.title("Ratings Distribution")
plt.show()

sns.countplot(y='listed_in', data=df, order=df['listed_in'].value_counts().index[:10])
plt.title("Top Genres")
plt.show()

sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

### Insights
- Movies are more common than TV shows.
- The United States produces the most content.
- Content production increased after 2010.
- Some genres are more popular than others.
- Most content falls under a few common ratings.
